## 실습 2 객실번호 지우고 데이터에서 NaN값 제거

```python
del df['Cabin']
df = df.dropna()
df = df.reset_index(drop = True)
df
```

## 실습 3 데이터에 Human_classfi 컬럼 추가하기

```python
df['Human_classif'] = None
df
```

```python
df.loc[(df.Sex == 'male') & (df.Age < 19), 'Human_classfi'] = 'Child_m'
df.loc[(df.Sex == 'female') & (df.Age < 19), 'Human_classfi'] = 'Child_f'
df.loc[(df.Sex == 'male') & (df.Age > 19), 'Human_classfi'] = 'Mr'
df.loc[(df.Sex == 'female') & (df.Age > 19), 'Human_classfi'] = 'Ms'
```

```python
df
```

## 실습 4 Human_classfi에 대한 설명이 담긴 데이터 프레임을 다음과 같이 생성하라. 그리고 원본 데이터프레임에 설명 데이터프레임을 조인하라

```python
data_dict = {
    'Human_classfi' : ['Child_m', 'Child_f', 'Mr', 'Ms'],
    'HC_DES' : ['19세 미만 남성', '19세 미만 여성', '성인 남성', '성인여성']
}
des_df = pd.DataFrame(data_dict)
des_df
```

```python
df2 = pd.merge(df, des_df, on = 'Human_classfi')
df2
```
