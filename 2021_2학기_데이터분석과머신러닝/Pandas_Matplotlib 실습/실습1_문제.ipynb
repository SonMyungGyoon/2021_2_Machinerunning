```python
df1 = pd.read_csv('covid_1.csv')
df2 = pd.read_csv('covid_2.csv')
df3 = pd.read_csv('who.csv')
```

```python
print(df1.columns)
print(df2.columns)
print(df3.columns)
```

```python
df1.head()
```

```python
df2.head()
```

```python
df = pd.concat([df1, df2])
df = df.reset_index(drop = True)
df
```

```python
df.head()
```

```python
df_res = pd.merge(df3, df, left_on='Country', right_on='Country')
df_res.head()
```

```python
df_res['TravelRisk'] = df_res['Death']/df_res['Checked']
df_res.head()
```

```python
df_res.describe()
```

```python
df_res['Class'] = 0
```

```python
df_res.loc[df_res['TravelRisk']<= 0.009406, 'Class'] = 0
df_res.loc[df_res['TravelRisk']> 0.009406, 'Class'] = 1
df_res.loc[df_res['TravelRisk']< 0.021521, 'Class'] = 2
```

```python
df_res
```

```python
df_res.to_csv('covid19.csv', sep = ',', encoding = 'utf8', index = False)
```
