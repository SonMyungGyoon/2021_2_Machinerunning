```python
df = pd.read_csv('파키슨병데이터.csv')
print(df.columns)
df.head()
```

```python
X = df[df.columns[:-1]].values
y = df[df.columns[-1]].values
X = X.astype(np.float64)
y = y.astype(np.float64)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0, shuffle=True)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Decision Tree로 진행

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from subprocess import check_call
```

```python
# decision tree 모형 인스턴스 생성 및 하이퍼파라미터(depth) = 3
tree_model = DecisionTreeClassifier(random_state=0, max_depth=3)
tree_model.fit(X_train, y_train)

print("훈련 세트 정확도: {:.3f}".format(tree_model.score(X_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(tree_model.score(X_test, y_test)))
print("특성 중요도:\n{}".format(tree_model.feature_importances_))
```
