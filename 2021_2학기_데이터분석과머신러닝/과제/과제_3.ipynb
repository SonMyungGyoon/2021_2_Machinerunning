```python
df = pd.read_csv('파키슨병데이터.csv')
print(df.columns)
df.head()
```

```python
X = df[df.columns[:-1]].values
y = df[df.columns[-1]].values
X = X.astype(np.float64)
y = y.astype(np.float64)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0, shuffle=True)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#Keras를 이용해서 분류

```python
import numpy as np
from tensorflow import keras
from keras import models
from keras import layers
```

https://tykimos.github.io/2019/05/10/KerasTuner/
하이퍼 파라미터 설정

drop out 값을 0, 0.1, 0.2,로 설정하교 결과를 비교, dropout을 0.1로 고정하고 학습률을 0.0001, 0.001, 0.005, 0.01로 바꿔서 결과를 비교

데이터결과가 0, 1 두가지로 나오므로 softmax함수가 아니라 sigmoid 함수를 사용하여 진행

drop out 사이즈가 0.1이고 은닉층이 12, 8로 이루어진 2계층과 softmax로 나오는 결과층을 가지는 모델 생성

```python
model = keras.Sequential()
model.add(keras.layers.Dense(12, input_dim = 22, activation = 'relu'))
model.add(keras.layers.Dense(8, activation = 'relu'))
model.add(keras.layers.Dense(1, activation = 'sigmoid'))

model.summary()
```

```python
keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = 'accuracy')
```

```python
model.fit(X_train, y_train, epochs=1000, batch_size=64, verbose=1, validation_data=(X_test ,y_test))
```

```python
scores = model.evaluate(X_train, y_train)
print("%s: %.2f%%" %(model.metrics_names[1], scores[1]*100))
scores = model.evaluate(X_test, y_test)
print("%s: %.2f%%" %(model.metrics_names[1], scores[1]*100))
```
