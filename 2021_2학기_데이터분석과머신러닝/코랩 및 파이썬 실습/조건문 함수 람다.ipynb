##가위바위보게임

```python
import random
user = int(input("가위(1) 바위(2) 보(3)를 입력!"))
computer = random.randint(1,3)

if user == computer:
  print("비겼습니다!")
elif user - computer == 1 or user - computer == -2:
  print("이겼습니다!")
else:
  print("졌습니다!")
```

##함수실습 : 두개의 정수 입력후 산술 연산 결과 반환

```python
def calc( a , b ):
  print(f"{a} + {b} = {a+b}")
  print(f"{a} - {b} = {a-b}")
  print(f"{a} * {b} = {a*b}")
  print(f"{a} ** {b} = {a**b}")
  print(f"{a} / {b} = {a/b}")
  print(f"{a} // {b} = {a//b}")
  print(f"{a} % {b} = {a%b}")
```

```python
a, b = input("두 숫자 입력 : ").split(" ")
a = int(a)
b = int(b)
```

```python
calc(a, b)
```

```python
def factorial(n):
  if n == 1:
    return 1
  else:
    return n*factorial(n-1)

factorial(5)
```

```python
lambda_factorial = lambda x : x*lambda_factorial(x-1) if x > 1 else 1
```

```python
lambda_factorial(5)
```
