```python
class Student:
  def __init__(self, name, studentNum, grade):
    self.__name = name
    self.__studentNum = studentNum
    self.__grade = grade

  def set_name(self, name):
    self.__name = name
  def set_number(self, studentNum):
    self.__studentNum = studentNum
  def set_grade(self, grade):
    self.__grade = grade
  def getName(self):
    return self.__name
  def getStudentNum(self):
    return self.__studentNum
  def getGrade(self):
    return self.__grade
  def printState(self):
    print(f"이름 : {name}")
    print(f"학번 : {studentNum}")
    print(f"학점 : {grade}")
  def __repr__(self):
    return "이름 : " + self.__name + "학번 :" + str(self.__studentNum) + "학점 : " + str(self.__grade)
```

```python
a = Student("손명균", 21720893, 3.5)
```
